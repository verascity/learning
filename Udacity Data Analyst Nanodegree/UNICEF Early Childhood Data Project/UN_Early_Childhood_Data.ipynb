```
#import codes and regions for comparison & mapping purposes
url2 = "https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/master/all/all.csv"
countries = pd.read_csv(url2, usecols=["name", "alpha-3", "region", "sub-region"])
  
#converting names in df to correspond with countries
df["Countries"].replace({'Democratic Republic of the Congo':'Congo (Democratic Republic of the)',
                         'Republic of Moldova':'Moldova (Republic of)', 
                         'State of Palestine':'Palestine, State of', 
                         'Swaziland':'Eswatini',
                        'The former Yugoslav Republic of Macedonia':'Macedonia (the former Yugoslav Republic of)'}, inplace=True)
```

```
#join region names and country codes to df -- more granular info than on the UN doc
joined_df = df.merge(countries, how="left", left_on="Countries", right_on="name")
print(joined_df)
```

```
#fixing weird issue with one country not being assigned a region
joined_df.loc[14, "name":"sub-region"] = joined_df.loc[14, "name":"sub-region"].fillna({"name" : "Central African Republic",
         "alpha-3": "CAF",
         "region": "Africa",
         "sub-region": "Sub-Saharan Africa"})

display(joined_df)
```

```
by_region = joined_df.groupby(["region"])
mean_by_region = by_region.mean()
print(mean_by_region)

by_subregion = joined_df.groupby(["sub-region"])
mean_by_subregion = by_subregion.mean()
#for r in by_subregion:
#  print(r)
  
print(mean_by_subregion)
print(by_subregion.std())
```

```
#checking quintiles against totals
def get_ratio(row):
  return row["Poorest 20%"]/row["Richest 20%"]

df["Ratio"] = df.apply(get_ratio, axis=1) #A higher ratio means less difference between rich/poor
df.corr() #Shows a weak correlation between ratio/total -- i.e. low totals may but don't necessarily mean low ratios

#In other words, countries with little overall support for learning may be only slightly more likely to have a sharper divide between rich/poor.

print(df.head(15))
print(df.corr())
```
